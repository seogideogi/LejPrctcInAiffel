* Pandas 라이브러리에서 subset 파라미터를 사용하는 함수들

dropna(subset=...): 지정된 열에 대해 결측값이 있는 행을 삭제합니다.

duplicated(subset=...): 지정된 열을 기준으로 중복 행을 식별합니다.

drop_duplicates(subset=...): 지정된 열을 기준으로 중복 행을 삭제합니다.

sort_values(by=...): 주어진 열을 기준으로 데이터 프레임을 정렬합니다. 여기서 by는 subset과 유사한 용도로 사용됩니다.

pd.read_csv(usecols=...): CSV 파일에서 특정 열만 읽습니다. usecols는 subset

pivot_table(values=..., index=..., columns=...): 피벗 테이블을 생성할 때 특정 열을 선택합니다. values, index, columns가 subset과 유사하게 작용할 수 있습니다.

merge(on=..., how=..., subset=...): 두 데이터 프레임을 병합할 때 설정할 수 있으며, on은 어떤 열을 기준으로 병합할지를 지정합니다.

# 5-4 평가(Rating) 관련 피처 생성 및 활용

```python
o_important = []
o_score = []
i_important = []
i_score = []

for i in dating_df.columns :
    if i.startswith('o_important') :
        o_important.append(i)
    elif i.startswith('o_score') :
        o_score.append(i) 
    elif i.startswith('i_important') :
        i_important.append(i)
    elif i.startswith('i_score') :
        i_score.append(i)
```

중요치 않아서 0을 준 경우. -99로 대체하기

```python
##  impotant 와 score 엮어서 유의미 데이터 만들기
dating_df['o_important_attractive'] * dating_df['o_score_attractive'] # 총 6개 항목이므로 일일이 하려면 12번임 -_-;
## 주의 -99 결측치 채운것. 결과도 -99 갖도록 함.
```

```python
def rating(data, important , score) :
    if data[score] == -99 :
        return -99
    elif data[important] == -99 :
        return -99
    else :
        return data[important] * data[score]
```

```python
dating_df.apply(lambda x:rating(x, 'o_important_attractive', 'o_score_attractive') , axis = 1) # x는 dating_df 자체 네
```

상대방의 rating  total , 나의 rating total 에 어떤 연산을 가해... 의미있는 값 도출 하기
상대방과 내가 서로 높은 점수를 줬다면, 매칭될 확률이 더 높을 것이므로.

< 예시 >

a사람 = 100 받음 : b사람 = 100 받음


c사람 = 50 받음 : d사람 = 150 받음


저 두 커플 의 점수의 평균을 둘다 100 이지만 , c-d 커플이 비매칭 될 확률이 더 높다.

< sol > 조합평균으로 평균 도출하기.
두 값이 유사할 수록 평균이 높게 나옴

```python
a = 100
b = 100

c = 50
d = 150
```

```python
(a+b)/2  # 산술평균
```

```python
(c+d)/2  # 산술평균
```

2ab / (a+b)  <- 조합평균

```python
2*a*b / (a+b)
```

```python
2*c*d / (c+d)
```
