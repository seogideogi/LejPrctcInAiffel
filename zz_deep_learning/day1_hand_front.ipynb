# 딥러닝 네트워크 설계하기
## Sequential Model 사용
tf.keras의 Sequential API를 이용하여 LeNet이라는 딥러닝 네트워크를 설계한 예입니다. 

8줄밖에 안되는 간단한 코드이지만, 손글씨 숫자 분류기를 구현하는 데는 충분합니다.

```python
model=keras.models.Sequential()
model.add(keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(28,28,1)))
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Conv2D(32, (3,3), activation='relu'))
model.add(keras.layers.MaxPooling2D((2,2)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(32, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))

print('Model에 추가된 Layer 개수: ', len(model.layers))
```

model=keras.models.Sequential()  

이 코드는 Sequential 모델을 초기화합니다. Sequential 모델은 층(layer)을 순차적으로 쌓을 수 있는 단순한 선형 스택입니다.
model.add(keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(28,28,1)))  

2D 합성곱 층(Convolutional layer)을 추가합니다. 필터의 수는 16개이며, 필터 크기는 3x3입니다. 활성화 함수로는 'relu'를 사용하고, 입력 데이터의 형상은 28x28의 단일 채널(흑백 이미지)을 가집니다. 컬러 표현을 하고 싶다면 3이라 입력 해야 함.
model.add(keras.layers.MaxPool2D(2,2))  

최대 풀링 층(Max pooling layer)을 추가합니다. 풀링 크기는 2x2로 설정되어 있어, 이미지의 해상도를 줄이는 역할을 합니다.
model.add(keras.layers.Conv2D(32, (3,3), activation='relu'))  

또 다른 2D 합성곱 층을 추가합니다. 이번에는 필터의 수가 32개로 늘어났으며, 필터 크기는 여전히 3x3입니다. 활성화 함수는 'relu'입니다.
model.add(keras.layers.MaxPooling2D((2,2)))  

두 번째 최대 풀링 층을 추가합니다. 크기는 역시 2x2입니다.
model.add(keras.layers.Flatten())  

평탄화 층(Flatten layer)을 추가하여 다차원 배열을 1차원 벡터로 변환합니다. 이는 전결층(Dense layer)과의 연결을 위해 필요합니다.
model.add(keras.layers.Dense(32, activation='relu'))  

Dense(완전 연결) 층을 추가합니다. 이 층은 32개의 뉴런을 가지며 활성화 함수로 'relu'를 사용합니다.
model.add(keras.layers.Dense(10, activation='softmax'))  

출력층으로 Dense 층을 추가합니다. 이 층은 10개의 뉴런을 가지며, 각 뉴런은 클래스 확률을 나타냅니다. 활성화 함수로는 'softmax'를 사용하여 출력값을 확률로 변환합니다. 일반적으로 분류 작업에서 사용합니다.


------------------

relu는 "Rectified Linear Unit"의 약자로, 인공 신경망의 활성화 함수 중 하나입니다. relu 함수는 입력값이 양수인 경우에는 그대로 출력하고, 음수인 경우에는 0을 출력하는 함수입니다. 수식으로 표현하면 다음과 같습니다:

[ \text{ReLU}(x) = \max(0, x) ]

relu의 주요 장점은 계산이 단순하고, 다른 활성화 함수에 비해 경사 하강법을 통한 학습이 더 빠르게 수렴할 수 있다는 점입니다. 또한, relu는 입력값이 음수일 때의 출력이 0이기 때문에 일부 뉴런을 "죽이도록" 해드롭아웃과 비슷하게 작용하며 과적합을 방지하는 데 도움이 될 수 있습니다. 이러한 이유로 딥러닝 모델에서 자주 사용되는 활성화 함수입니다.



* Conv2D 레이어의 첫 번째 인자는 사용하는 이미지 특징의 수입니다. 여기서는 16과 32를 사용했습니다. 가장 먼저 16개의 이미지 특징을, 그 뒤에 32개의 이미지 특징씩을 고려하겠다는 뜻입니다. 우리의 숫자 이미지는 사실 매우 단순한 형태의 이미지입니다. 만약 강아지 얼굴 사진이 입력 이미지라면 훨씬 디테일하고 복잡한 영상일 것입니다. 그럴 경우에는 이 특징 숫자를 늘려주는 것을 고려해 볼 수 있습니다.
* Dense 레이어의 첫 번째 인자는 분류기에 사용되는 뉴런의 숫자 입니다. 이 값이 클수록 보다 복잡한 분류기를 만들 수 있습니다. 10개의 숫자가 아닌 알파벳을 구분하고 싶다면, 대문자 26개, 소문자 26개로 총 52개의 클래스를 분류해 내야 합니다. 그래서 32보다 큰 64, 128 등을 고려해 볼 수 있을 것입니다.
* 마지막 Dense 레이어의 뉴런 숫자는 결과적으로 분류해 내야 하는 클래스 수로 지정하면 됩니다. 숫자 인식기에서는 10, 알파벳 인식기에서는 52가 되겠지요.

# 딥러닝 네트워크 학습시키기

우리가 만든 네트워크의 입력은 (데이터갯수, 이미지 크기 x, 이미지 크기 y, 채널수) 와 같은 형태를 가집니다. 이전 스텝에서 첫 번째 레이어에 input_shape=(28,28,1)로 지정했던 것을 기억하시나요?
그런데 print(x_train.shape) 을 해보면, (60000, 28, 28) 로 채널수에 대한 정보가 없습니다. 따라서 (60000, 28, 28, 1) 로 만들어 주어야 합니다 (여기서 채널수 1은 흑백 이미지를 의미합니다. 컬러 이미지라면 R, G, B 세 가지 값이 있기 때문에 3이겠죠?).

```python
print(f"Before Reshape - x_train_norm shape: {x_train_norm.shape}")
print(f"Before Reshape - x_test_norm shape: {x_test_norm.shape}")

x_train_reshaped=x_train_norm.reshape( -1, 28, 28, 1)  # 데이터갯수에 -1을 쓰면 reshape시 자동계산됩니다.
x_test_reshaped=x_test_norm.reshape( -1, 28, 28, 1)

print(f"After Reshape - x_train_reshaped shape: {x_train_reshaped.shape}")
print(f"After Reshape - x_test_reshaped shape: {x_test_reshaped.shape}")
```

데이터를 reshape하는 이유는 신경망 모델에 입력할 때 데이터를 올바른 형상으로 정렬하여 모델이 기대하는 형상과 일치시킴으로써, 모델의 계층이 데이터를 올바르게 처리할 수 있도록 하기 위함입니다.

특히, 이미지 데이터의 경우:

형상 일치: 이미지 데이터는 일반적으로 특정 형상을 갖습니다. 이 형상은 (높이, 너비, 채널 수)로 표현됩니다. 예를 들어, 28x28의 흑백 이미지는 (28, 28, 1) 형상을 가집니다. reshape를 통해 데이터의 차원을 이와 같이 맞춰줌으로써 모델이 정확히 데이터를 인식하게 합니다.

배치 차원 추가: -1을 사용하여 첫 번째 차원을 설정하는 것은 데이터의 개수(배치크기)를 의미합니다. 모델은 한 번에 여러 이미지를 처리할 수 있도록 미니배치 형식으로 데이터를 받습니다. -1은 전체 데이터 개수에 맞춰 자동으로 배치차원을 계산하여 설정합니다.

채널 차원 추가: 많은 이미지 데이터는 height x width 형식으로 일차원일 수 있습니다. reshape를 통해 (height, width, 1)로 변환하여 채널 차원을 추가합니다. 이는 특히 콘벌루션 레이어가 데이터를 다루기 쉽게 해줍니다.

따라서, reshape는 데이터를 모델이 요구하는 형상으로 변경하여 신경망이 이를 올바르게 처리하고 학습할 수 있도록 해줍니다.


----------------


x_train 학습 데이터로 딥러닝 네트워크를 학습시켜 봅시다. 여기서 epochs=10 은 전체 60,000개의 데이터를 10번 반복 사용해서 학습을 시키라는 뜻입니다. 물론 model의 입력 정의에 형태를 맞춘 x_train_reshaped가 사용되어야겠죠.

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

model.fit(x_train_reshaped, y_train, epochs=10)
```

Epoch  최대 0.9960

# 얼마나 잘 만들었는지 확인하기

```python
test_loss, test_accuracy = model.evaluate(x_test_reshaped, y_test, verbose=2)
print(f"test_loss: {test_loss}")
print(f"test_accuracy: {test_accuracy}")
```

## 어떤 데이터를 잘못 추론 했을까? 눈으로 확인 하기

```python
predicted_result = model.predict(x_test_reshaped)  # model이 추론한 확률값. 
predicted_labels = np.argmax(predicted_result, axis=1)

idx=0  #1번째 x_test를 살펴보자. 

print('model.predict() 결과 : ', predicted_result[idx])
print('model이 추론한 가장 가능성이 높은 결과 : ', predicted_labels[idx])
print('실제 데이터의 라벨 : ', y_test[idx])
```

model.evaluate() 대신 model.predict()를 사용하면 model이 입력값을 보고 실제로 추론한 확률분포를 출력할 수 있습니다. 우리가 만든 model이란 사실 10개의 숫자 중 어느 것일지에 대한 확률값을 출력하는 함수입니다. 이 함수의 출력값, 즉 확률값이 가장 높은 숫자가 바로 model이 추론한 숫자가 되는 거죠.



정말 숫자 7인가? 확인


```python
plt.imshow(x_test[idx],cmap=plt.cm.binary)
plt.show()
```

model이 추론해 낸 숫자와 실제 라벨의 값이 다른 경우는?

```python
import random
wrong_predict_list=[]
for i, _ in enumerate(predicted_labels):
    # i번째 test_labels과 y_test이 다른 경우만 모아 봅시다. 
    if predicted_labels[i] != y_test[i]:
        wrong_predict_list.append(i)

# wrong_predict_list 에서 랜덤하게 5개만 뽑아봅시다.
samples = random.choices(population=wrong_predict_list, k=5)

for n in samples:
    print("예측확률분포: " + str(predicted_result[n]))
    print("라벨: " + str(y_test[n]) + ", 예측결과: " + str(predicted_labels[n]))
    plt.imshow(x_test[n], cmap=plt.cm.binary)
    plt.show()
```

# 더 좋은 네트워크 만들어 보기

네트워크의 구조 자체는 바꾸지 않으면서도 우리가 해볼 수 있는 것들.


- 딥러닝 네트워크 설계하기에서 살펴본 하이퍼파라미터 조정.
- Conv2D 레이어에서 입력 이미지의 특징 수를 늘리거나 줄여 보거나, Dense 레이어에서 뉴런수를 바꾸어 보거나, 학습 반복 횟수인 epoch 값을 변경해 볼 수 있다.

```python
# 바꿔 볼 수 있는 하이퍼파라미터들
# n_channel_1=16
# n_channel_2=32
# n_dense=32
# n_train_epoch=10

# 313/313 - 2s - loss: 0.0420 - accuracy: 0.9888
# test_loss: 0.041966501623392105 
# test_accuracy: 0.9887999892234802
##------------------------------------------------------
# n_channel_1=32
# n_channel_2=32
# n_dense=32
# n_train_epoch=10

# 313/313 - 3s - loss: 0.0383 - accuracy: 0.9886
# test_loss: 0.03826409950852394 
# test_accuracy: 0.9886000156402588
#--------------------------------------------------------
# n_channel_1=16
# n_channel_2=32
# n_dense=32
# n_train_epoch=5

# 313/313 - 2s - loss: 0.0447 - accuracy: 0.9860
# test_loss: 0.04465566575527191 
# test_accuracy: 0.9860000014305115
    
    
n_channel_1=32
n_channel_2=64
n_dense=64
n_train_epoch=10

# 313/313 - 3s - loss: 0.0308 - accuracy: 0.9922
# test_loss: 0.030795887112617493 
# test_accuracy: 0.9922000169754028

model=keras.models.Sequential()
model.add(keras.layers.Conv2D(n_channel_1, (3,3), activation='relu', input_shape=(28,28,1)))
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Conv2D(n_channel_2, (3,3), activation='relu'))
model.add(keras.layers.MaxPooling2D((2,2)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(n_dense, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))

model.summary()
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

# 모델 훈련
model.fit(x_train_reshaped, y_train, epochs=n_train_epoch)

# 모델 시험
test_loss, test_accuracy = model.evaluate(x_test_reshaped, y_test, verbose=2)
print(f"test_loss: {test_loss} ")
print(f"test_accuracy: {test_accuracy}")
```
