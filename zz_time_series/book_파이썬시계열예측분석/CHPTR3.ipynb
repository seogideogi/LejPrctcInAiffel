# 3.1 확률보행 ( random walk )
* 무작위 상승 or 하락이 발생할 확률이 동일한 프로세스
* 일반적으로 일일종가와 같은 금융 및 경제 데이터에서 관찰 됨.
* 양의 추세 or 음의 추세가 오래 지속되는 기간을 포함
* 급격한 방향 전환을 동반 하기도 함


* 현재값 : $y_t$
* 이전시간 단계 값 : $y_{t-1}$
* 상수 : $C$
* 백색소음 or 난수
    : $\epsilon_t$
    (분산이 1이고 평균이 0인 표준정규분포를 보임)
    
    
    
확률보행

$y_t$ = $C$ + $y_{t-1}$ + $\epsilon_t$

$C$가 0이 아닌 경우, 표류(drift)가 있는 확률 보행 이라 함

## 3.1.1 확률보행 프로세스 시뮬레이션
* 확률보행 프로세스를 0으로 초기화
* 상수 C = 0 설정하면
* t시점의 값은 t=1 부터 t 시점까지의 백색 소음의 합 임.   p. 38

```python
import numpy as np

np.random.seed(42)

steps = np.random.standard_normal(1000)
```

```python
steps
```

```python
steps[0] = 0 # 수열 첫번재 값 0으로 초기화
```

```python
steps
```

```python
random_walk = np.cumsum(steps)
random_walk
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot(random_walk)
ax.set_xlabel('Timesteps')
ax.set_ylabel('Value')
plt.tight_layout()
```
